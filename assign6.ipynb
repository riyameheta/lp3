Implement K-Means clustering/ hierarchical clustering on sales_data_sample.csv dataset.
Determine the number of clusters using the elbow method.
Dataset link : https://www.kaggle.com/datasets/kyanyoga/sample-sales-data

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('sales_data_sample.csv', encoding='ISO-8859-1')
```

```python
df = df.drop(columns=['ADDRESSLINE1', 'ADDRESSLINE2', 'CITY', 'STATE', 'POSTALCODE', 'TERRITORY', 'CONTACTLASTNAME', 'CONTACTFIRSTNAME', 'PHONE', 'CUSTOMERNAME', 'ORDERNUMBER', 'ORDERDATE', 'STATUS', 'QTR_ID', 'MONTH_ID', 'YEAR_ID'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
categories = ['PRODUCTLINE', 'PRODUCTCODE', 'COUNTRY', 'DEALSIZE']
le = LabelEncoder()
```

```python
for col in categories:
    df[col] = le.fit_transform(df[col])
```

```python
data = StandardScaler().fit_transform(df)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range (1,15):
    kmeans = KMeans(n_clusters=k, init=('k-means++'), random_state=15)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
```

```python
k = list(range(1,15))
plt.plot(k,wcss)
plt.title('WCSS vs Number of clusters')
plt.xlabel('clusters')
plt.ylabel('scores')
plt.grid()
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(data, method='ward')
dendrogram(linked, orientation='top')
```
